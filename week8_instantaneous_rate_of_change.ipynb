Last time we considered average change of rate by computing rise over run.

Question: what is a change that average change is more helpful than instantaneous rate of change? What is a change that instantaneous rate of change is more useful than average rate of change?

Consider the function $y=\sin(x)$. We will study the rate of change at several values of $x$. At each point of $x$, we make the run shorter, and observe the change of rate changing.

The points are labeled $x0$, and we use $h$ to denote run. The formula will be

rise over run at $x_0$ with run at $h$ is

$\frac{f(x_0 + h) - f(x_0)}{(x_0 + h) - x_0} = \frac{f(x_0 + h) - f(x_0)}{h}$

```python
def rise_over_run_sine(h, x0):

  rise = np.sin(x0 + h) - np.sin(x0)
  run = h

  return rise / run
```

```python
# test
rise_over_run_sine(0.1, 0)
```

We compute a sequence of rise over run, as h becomes smaller and smaller, and save the result in a list

```python
ror_sine_list = []
h_list = []

for j in range(50):
  htemp = 10**(-j) # htemp = 10^(-j)
  h_list.append(htemp)
  ror_sine_list.append(rise_over_run_sine(htemp, 0))

n = np.linspace(0, 50, 50)
```

```python
plt.figure()
plt.plot(n, ror_sine_list, label='rise over run values', color='b')
plt.plot(n, h_list, label='h values', color='r')
plt.xlabel('n')

plt.legend()
# plt.axis('equal')
# plt.grid(True)
plt.show()
```

The plot the line $y - \sin(x_0) = (rise\_over\_run\_sine(h, x_0)) (x - x_0)$ with the curve $y=\sin(x)$.

In this case, as we take $x_0=0$, so the above line is $y = (rise\_over\_run\_sine(h, 0))\ x$

```python
x = np.linspace(-np.pi/2, np.pi/2, 50)
sinx = np.sin(x)

h = 0.1
x0 = 0
y = rise_over_run_sine(h, x0) * (x - x0) + np.sin(x0)
```

```python
plt.figure()
plt.plot(x, y, label='rise over run values', color='b')
plt.plot(x, sinx, label='$y=\\sin(x)$', color='r')
plt.scatter(x0, np.sin(x0), label = 'point $x_0, y_0$')
plt.xlabel('n')

plt.legend()
plt.show()
```
