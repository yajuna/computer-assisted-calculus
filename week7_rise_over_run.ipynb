An example is, if a kid grows 2 inches in a month, we say this kid grows _faster_ than a kid who grows 1 inch in a month. We formulate this problem in math by considering rise over run.

### As an example, we consider [these two sets of data](https://www.tylervigen.com/spurious/correlation/5898_annual-us-household-spending-on-pork_correlates-with_lululemons-stock-price-lulu) from Spurious Correlations:
Annual US household spending on pork vs. Lululemon's stock price.


```python
n = 2022 - 2008 + 1
year = np.linspace(2008, 2022, n)
spending_on_pork = np.array([163,168,149,162,166,170,177,165,169,181,180,187,214,223,246,])
lululemon_stock_price = np.array([23.7,3.98,15.44,34.7,47.5,77.73,59.08,56.01,53.71,65.94,77.31,118.89,232.9,351.67,392.2,])
```

Visualize data as functions of year (time).

```python
plt.figure()
plt.plot(year, spending_on_pork, label='Household Spending', color='b')
plt.plot(year, lululemon_stock_price, label='Lululemon stock', color='r')
plt.xlabel('Year')

plt.legend()
# plt.axis('equal')
# plt.grid(True)
plt.show()
```

### We describe how fast something changes by looking at the change (increase or decrease) within a time period.

We define rate of change = rise / run = change in value / time elapsed

We can use this to estimate the _average change_ in a given time period. Here we compute the average change of spending on pork between year 2008 and 2010, as well as Lululemon stock price between 2020 to 2022.

```python
ave_spend_change = (spending_on_pork[2] - spending_on_pork[0]) / (2 - 0)
ave_stock_change = (lululemon_stock_price[-1] - lululemon_stock_price[-3]) / 2
```
